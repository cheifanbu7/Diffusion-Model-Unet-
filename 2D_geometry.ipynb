```python
# Import necessary modules
import numpy as np
from pint import Quantity
from ansys.geometry.core import launch_modeler
from ansys.geometry.core.math import Plane, Point3D, Point2D, UnitVector3D
from ansys.geometry.core.misc import UNITS
from ansys.geometry.core.sketch import Sketch
```

```python
# Create a new sketch object
sketch = Sketch()

# Define the rectangular plate and circular features
(sketch.segment(Point2D([0, 0], unit=UNITS.mm), Point2D([10, 0], unit=UNITS.mm))
    .segment_to_point(Point2D([10, 10], unit=UNITS.mm))
    .segment_to_point(Point2D([0, 10], unit=UNITS.mm))
    .segment_to_point(Point2D([0, 0], unit=UNITS.mm))
)
```

```python
# Launch SpaceClaim modeler in hidden mode
modeler = launch_modeler("spaceclaim", hidden=True)

# Create a new design named "2D_Plate"
design = modeler.create_design("2D_Plate")
```

```python
# Create a 2D surface body from the closed sketch
body = design.create_surface("PlateLayer", sketch)
```

```python
# Create a named selection for the surface body
plate_named_selection = design.create_named_selection("Plate", bodies=[body])
```

```python
# Plot the design
design.plot()
```

```python
# Export the annotated design to a PMDB file in the current working directory.
pmdb_path = design.export_to_pmdb()
print(f"Design exported to: {pmdb_path}")

```

```python
# Close the modeler session 
modeler.close()
```
